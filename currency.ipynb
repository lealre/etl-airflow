```python
import pandas_datareader.data as web
import datetime
from pprint import pprint
import pandas as pd
```

```python
start = datetime.datetime(2022, 4, 5)
end = datetime.datetime(2024, 4, 10)

usd_yen = web.DataReader('DEXJPUS', 'fred', start, end)
eur_usd = web.DataReader('DEXUSEU', 'fred', start)

eur_usd = eur_usd.resample('M').last()
eur_usd.index = eur_usd.index.strftime('%Y-%m')

usd_yen = usd_yen.resample('M').last()
usd_yen.index = usd_yen.index.strftime('%Y-%m')
yen_usd = (1/ usd_yen) #normalize

df_conversion_rates = pd.concat([yen_usd, eur_usd], axis = 1)

print(df_conversion_rates)
print(eur_usd.head())
```
